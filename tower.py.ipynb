```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
t, v= np.loadtxt("droptower_vdata.txt", unpack = True) #reads times and velocities from the given data file 

#Plots three separate graphs

#Plots position
initial_pos = 30
x = integrate.cumtrapz(v, t, initial=0) + initial_pos
plt.plot(t, x)
plt.suptitle('Position', fontsize=14, fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.show()

#Plots velocity
plt.plot(t,v) #plots the velocity
plt.suptitle('Velocity', fontsize=14, fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.show()

#Plots acceleration
a = np.diff(v)/t[1:11]
plt.plot(t[1:11], a) #plots the acceleration
plt.suptitle('Acceleration', fontsize=14, fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s$^2$)')
plt.show()


```

```python
#Plots three curves on one graph

plt.plot(t,x,color='red',linestyle='dotted', lw=4, marker='o', label = 'position')
plt.plot(t,v,color='blue',linestyle='dashed', lw=4, marker='o', label = 'velocity')
plt.plot(t[1:11], a, color='green', lw=4, marker='o', label = 'acceleration')
plt.suptitle('Tower Ride Data', fontsize=14, fontweight='bold')
plt.legend()
plt.show()
```

```python
#Plots three curves as sub-plots of one graphs
f, (ax0,ax1,ax2) = plt.subplots(3, sharey=True)
ax0.plot(t,x,color='r', label= 'position')
ax1.plot(t,v,color='b', label = 'velocity')
ax2.plot(t[1:11],a,color='g', label = 'acceleration')
f.subplots_adjust(hspace=0.5)
plt.suptitle('Tower Ride Data', fontsize=12, fontweight='bold')
ax0.set_title('Position')
ax1.set_title('Velocity')
ax2.set_title('Acceleration')
plt.xlabel('Time (s)')
plt.show()
```
